```python
def redpr(str):
    print('\x1b[31m' + str + '\x1b[0m')
    
def greenpr(str):
    print('\x1b[32m' + str + '\x1b[0m')
    
def colorpr2(str0, str1, col1, str2, col2):
    print(str0 + '\x1b[' + col1 + 'm' + str1 + '\x1b[0m', '   at  ', '\x1b[' + col2 + 'm' + str2 + '\x1b[0m')
```

```python
week = 5
lastweek = week - 1
wk = str(week)
lwk = str(lastweek)

with open('schedule_2018.txt') as f:
    #passer = f.readlines()
    schedule = list(f)

with open('Week' + lwk + '/passingOFF_ydsper_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    passOFF = list(f)
    
with open('Week' + lwk + '/passingDEF_ydsper_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    passDEF = list(f)

with open('Week' + lwk + '/sacks_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    sacks = list(f)
    
with open('Week' + lwk + '/timessacked_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    times = list(f)
    
with open('Week' + lwk + '/rushingOFF_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    rushOFF = list(f)
    
with open('Week' + lwk + '/rushingDEF_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    rushDEF = list(f)
```

```python
for i in range(len(schedule)):
    if ( schedule[i][0]==wk and schedule[i][1]==' ' ) or schedule[i][0:2]==wk:
        
        team_away = schedule[i][3:29]
        team_home = schedule[i][28:]
        
        spread =  2.5
        
        for j in range(32):
            
            if team_away.lstrip().rstrip() in passOFF[j]:
                place1 = int(passOFF[j][0:2])
                passOFF_a_str = passOFF[j].rstrip()[-5:]
                passOFF_a = float(passOFF_a_str)
                col1 = '30'
                if place1 <= 10:
                    col1 = '32'
                elif place1 >= 23:
                    col1 = '31'
                    
                str1 = str(place1).rjust(3) + ' (' + passOFF_a_str.rjust(6) + ')'
               
            if team_home.lstrip().rstrip() in passOFF[j]:
                place2 = int(passOFF[j][0:2])
                passOFF_h_str = passOFF[j].rstrip()[-5:]
                passOFF_h = float(passOFF_h_str)
                col2 = '30'
                if place2 <= 10:
                    col2 = '32'
                elif place2 >= 23:
                    col2 = '31'
                    
                str2 = str(place2).rjust(3) + ' (' + passOFF_h_str.rjust(6) + ')'
         
            if team_away.lstrip().rstrip() in passDEF[j]:
                place1b = int(passDEF[j][0:2])
                passDEF_a_str = passDEF[j].rstrip()[-5:]
                passDEF_a = float(passDEF_a_str)
                col1b = '30'
                if place1b <= 10:
                    col1b = '32'
                elif place1b >= 23:
                    col1b = '31'
                    
                str1b = str(place1b).rjust(3) + ' (' + passDEF_a_str.rjust(6) + ')'
               
            if team_home.lstrip().rstrip() in passDEF[j]:
                place2b = int(passDEF[j][0:2])
                passDEF_h_str = passDEF[j].rstrip()[-5:]
                passDEF_h = float(passDEF_h_str)
                col2b = '30'
                if place2b <= 10:
                    col2b = '32'
                elif place2b >= 23:
                    col2b = '31'
                    
                str2b = str(place2b).rjust(3) + ' (' + passDEF_h_str.rjust(6) + ')'
        
            if team_away.lstrip().rstrip() in sacks[j]:
                place3 = int(sacks[j][0:2])
                sacks_a_str = sacks[j].rstrip()[-4:]
                sacks_a = float(sacks_a_str)
                col3 = '30'
                if place3 <= 10:
                    col3 = '32'
                elif place3 >= 23:
                    col3 = '31'
                    
                str3 = str(place3).rjust(3) + ' (' + sacks_a_str.rjust(6) + ')'
               
            if team_home.lstrip().rstrip() in sacks[j]:
                place4 = int(sacks[j][0:2])
                sacks_h_str = sacks[j].rstrip()[-4:]
                sacks_h = float(sacks_h_str)
                col4 = '30'
                if place4 <= 10:
                    col4 = '32'
                elif place4 >= 23:
                    col4 = '31'
                    
                str4 = str(place4).rjust(3) + ' (' + sacks_h_str.rjust(6) + ')'
                
            if team_away.lstrip().rstrip() in times[j]:
                place5 = int(times[j][0:2])
                times_a_str = times[j].rstrip()[-4:]
                times_a = float(times_a_str)
                col5 = '30'
                if place5 <= 10:
                    col5 = '32'
                elif place5 >= 23:
                    col5 = '31'
                    
                str5 = str(place5).rjust(3) + ' (' + times_a_str.rjust(6) + ')'
               
            if team_home.lstrip().rstrip() in times[j]:
                place6 = int(times[j][0:2])
                times_h_str = times[j].rstrip()[-4:]
                times_h = float(times_h_str)
                col6 = '30'
                if place6 <= 10:
                    col6 = '32'
                elif place6 >= 23:
                    col6 = '31'
                    
                str6 = str(place6).rjust(3) + ' (' + times_h_str.rjust(6) + ')'
            
            if team_away.lstrip() in rushOFF[j]:
                place7 = int(rushOFF[j][0:2])
                rushOFF_a_str = rushOFF[j].rstrip()[-5:]
                rushOFF_a = float(rushOFF_a_str)
                col7 = '30'
                if place7 <= 10:
                    col7 = '32'
                elif place7 >= 23:
                    col7 = '31'
                    
                str7 = str(place7).rjust(3) + ' (' + rushOFF_a_str.rjust(6) + ')'
               
            if team_home.lstrip().rstrip() in rushOFF[j]:
                place8 = int(rushOFF[j][0:2])
                rushOFF_h_str = rushOFF[j].rstrip()[-5:]
                rushOFF_h = float(rushOFF_h_str)
                col8 = '30'
                if place8 <= 10:
                    col8 = '32'
                elif place8 >= 23:
                    col8 = '31'
                    
                str8 = str(place8).rjust(3) + ' (' + rushOFF_h_str.rjust(6) + ')'
                
            if team_away.lstrip() in rushDEF[j]:
                place9 = int(rushDEF[j][0:2])
                rushDEF_a_str = rushDEF[j].rstrip()[-5:]
                rushDEF_a = float(rushDEF_a_str)
                col9 = '30'
                if place9 <= 10:
                    col9 = '32'
                elif place9 >= 23:
                    col9 = '31'
                    
                str9 = str(place9).rjust(3) + ' (' + rushDEF_a_str.rjust(6) + ')'
               
            if team_home.lstrip().rstrip() in rushDEF[j]:
                place10 = int(rushDEF[j][0:2])
                rushDEF_h_str = rushDEF[j].rstrip()[-5:]
                rushDEF_h = float(rushDEF_h_str)
                col10 = '30'
                if place10 <= 10:
                    col10 = '32'
                elif place10 >= 23:
                    col10 = '31'
                    
                str10 = str(place10).rjust(3) + ' (' + rushDEF_h_str.rjust(6) + ')'
        
        spread += (place1 - place2 
                   + place1b - place2b
                   + place3 - place4
                   + place5 - place6
                   + place7 - place8
                   + place9 - place10)/10
        
        print('-----------------------------------------------------------')
        print(team_away + '  {0:.1f}  at  '.format(spread) + team_home)
        print('PASSING:')
        colorpr2('ydsper offense:      ', str1, col1, str2, col2)
        colorpr2('ydsper defense:      ', str1b, col1b, str2b, col2b)
        print()
        print('LINE PLAY:')
        colorpr2('oline:               ', str5, col5, str6, col6)
        colorpr2('sacks:               ', str3, col3, str4, col4)
        print()
        print('RUSHING:')
        colorpr2('ydsper offense:      ', str7, col7, str8, col8)
        colorpr2('ydsper defense:      ', str9, col9, str10, col10)
    else:
        continue
```
