```python
def redpr(str):
    print('\x1b[31m' + str + '\x1b[0m')
    
def greenpr(str):
    print('\x1b[32m' + str + '\x1b[0m')
    
def colorpr2(strA, str1, col1, strB, str2, col2):
    print(strA + '\x1b[' + col1 + 'm' + str1.rjust(6) + '\x1b[0m    ', strB.ljust(16) + '\x1b[' + col2 + 'm' + str2.rjust(6) + '\x1b[0m')
```

```python
week = 10
lastweek = week - 1
wk = str(week)
lwk = str(lastweek)

with open('schedule_2018.txt') as f:
    #passer = f.readlines()
    schedule = list(f)

with open('Week' + lwk + '/SRS_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    SRS = list(f)
    
with open('Week' + lwk + '/penaltiesOFF_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    penalties = list(f)
    
with open('Week' + lwk + '/interceptions_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    interceptions = list(f)
    
with open('Week' + lwk + '/timessacked_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    timessacked = list(f)

with open('Week' + lwk + '/sacks_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    sacks = list(f)
    
with open('Week' + lwk + '/rushingOFF_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    rushOFF = list(f)
    
with open('Week' + lwk + '/rushingDEF_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    rushDEF = list(f)
    
with open('Week' + lwk + '/3DconvOFF_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    DconvOFF = list(f)
    
with open('Week' + lwk + '/3DconvDEF_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    DconvDEF = list(f)
    
with open('Week' + lwk + '/redzoneOFF_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    redzoneOFF = list(f)
    
with open('Week' + lwk + '/redzoneDEF_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    redzoneDEF = list(f)
```

```python
team_list = []
data_dict = {}

for i in range(len(SRS)):
    
    # SRS:
    team = SRS[i][:22].lstrip().rstrip()
    team_list.append(team)
    
    val = float(SRS[i].rstrip()[-7:])
    
    data_dict[team] = {}
    data_dict[team]['SRS'] = val
    
data_dict

    
for i in range(len(SRS)):
    # penalties (offense):
    team = penalties[i][2:25].lstrip()
    
    val1 = float(penalties[i].rstrip()[-3:])
    val2 = float(penalties[i].rstrip()[-11:-4])
    
    rnk = int(penalties[i][0:2].rstrip())
    
    data_dict[team]['penalties'] = [val1, val2]
    
    # interceptions:
    team = interceptions[i][2:25].lstrip()
    
    val = float(interceptions[i].rstrip()[-3:])
    
    rnk = int(interceptions[i][0:2].rstrip())
    
    data_dict[team]['interceptions'] = [val, rnk]
    
    # timessacked:
    team = timessacked[i][2:25].lstrip()
    
    val = float(timessacked[i].rstrip()[-3:])
    
    rnk = int(timessacked[i][0:2].rstrip())
    
    data_dict[team]['timessacked'] = [val, rnk]
    
    # sacks:
    team = sacks[i][2:25].lstrip()
    
    val = float(sacks[i].rstrip()[-3:])
    
    rnk = int(sacks[i][0:2].rstrip())
    
    data_dict[team]['sacks'] = [val, rnk]
    
    # rushing offense:
    team = rushOFF[i][2:25].lstrip()
    
    val = float(rushOFF[i].rstrip()[-5:])
    
    rnk = int(rushOFF[i][0:2].rstrip())
    
    data_dict[team]['rushOFF'] = [val, rnk]
    
    # rushing defense:
    team = rushDEF[i][2:25].lstrip()
    
    val = float(rushDEF[i].rstrip()[-5:])
    
    rnk = int(rushDEF[i][0:2].rstrip())
    
    data_dict[team]['rushDEF'] = [val, rnk]
    
    # 3rd down conversions offense:
    team = DconvOFF[i][2:25].lstrip()
    
    val = float(DconvOFF[i].rstrip()[-5:-1])
    
    rnk = int(DconvOFF[i][0:2].rstrip())
    
    data_dict[team]['3DconvOFF'] = [val, rnk]
    
    # 3rd down conversions defense:
    team = DconvDEF[i][2:25].lstrip()
    
    val = float(DconvDEF[i].rstrip()[-5:-1])
    
    rnk = int(DconvDEF[i][0:2].rstrip())
    
    data_dict[team]['3DconvDEF'] = [val, rnk]
    
    # redzone offense:
    team = redzoneOFF[i][2:25].lstrip()
    
    val = float(redzoneOFF[i].rstrip()[-5:-1])
    
    rnk = int(redzoneOFF[i][0:2].rstrip())
    
    data_dict[team]['redzoneOFF'] = [val, rnk]
    
    # redzone defense:
    team = redzoneDEF[i][2:25].lstrip()
    
    val = float(redzoneDEF[i].rstrip()[-5:-1])
    
    rnk = int(redzoneDEF[i][0:2].rstrip())
    
    data_dict[team]['redzoneDEF'] = [val, rnk]
    
    
team_list = sorted(team_list)
```

```python
pen_ar = []

for team in team_list:
    val0 = data_dict[team]['penalties'][0]
    val1 = data_dict[team]['penalties'][1]
    pen_ar.append([val0/val1, team] )

penper = sorted(pen_ar, reverse=False)
    
for i in range(32):
    team = penper[i][1]
    data_dict[team]['penalties'][0] = penper[i][0]
    data_dict[team]['penalties'][1] = i + 1
```

```python
SRS_ar = []
OFF_ar = []
DEF_ar = []

for team in team_list:
    SRS_ar.append([data_dict[team]['SRS'], team]) 
    rnk_1 = data_dict[team]['penalties'][1]
    rnk_3 = data_dict[team]['interceptions'][1]
    rnk_5 = data_dict[team]['timessacked'][1]
    rnk_9 = data_dict[team]['rushOFF'][1]
    rnk_13 = data_dict[team]['3DconvOFF'][1]
    rnk_17 = data_dict[team]['redzoneOFF'][1]
    
    OFF_val = np.mean([rnk_1, rnk_3, rnk_5, rnk_9, rnk_13, rnk_17])
    OFF_ar.append([OFF_val, team, rnk_1, rnk_3, rnk_5, rnk_9, rnk_13, rnk_17])
    
    rnk_7 = data_dict[team]['sacks'][1]
    rnk_11 = data_dict[team]['rushDEF'][1]
    rnk_15 = data_dict[team]['3DconvDEF'][1]
    rnk_19 = data_dict[team]['redzoneDEF'][1]
    
    DEF_val = np.mean([rnk_7, rnk_11, rnk_15, rnk_19])
    DEF_ar.append([DEF_val, team, rnk_7, rnk_11, rnk_15, rnk_19])

    
power_rankings = sorted(SRS_ar, reverse=True)
power_OFF = sorted(OFF_ar, reverse=False)
power_DEF = sorted(DEF_ar, reverse=False)


print('SRS:')
print()
for i in range(32):
    print('{:2d}'.format(i + 1), power_rankings[i][1].rjust(22), '{0:+5.1f}'.format(power_rankings[i][0] ))
    
print()
print('Offense:')
print()
for i in range(32):
    print('{:2d}'.format(i + 1), power_OFF[i][1].rjust(22), '{0:4.1f}'.format(power_OFF[i][0]), '   {}'.format(power_OFF[i][2:]))
    team = power_OFF[i][1].lstrip().rstrip()
    data_dict[team]['OFFENSE'] = i + 1
    
print()
print('Defense:')
print()
for i in range(32):
    print('{:2d}'.format(i + 1), power_DEF[i][1].rjust(22), '{0:4.1f}'.format(power_DEF[i][0]), '   {}'.format(power_DEF[i][2:]))
    team = power_DEF[i][1].lstrip().rstrip()
    data_dict[team]['DEFENSE'] = i + 1
```
