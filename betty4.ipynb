```python
import numpy as np
```

```python
def redpr(str):
    print('\x1b[31m' + str + '\x1b[0m')
    
def greenpr(str):
    print('\x1b[32m' + str + '\x1b[0m')
    
def colorpr2(str0, str1, col1, str2, col2):
    print(str0 + '\x1b[' + col1 + 'm' + str1.rjust(8) + '\x1b[0m', '\x1b[' + col2 + 'm' + str2.rjust(16) + '\x1b[0m')
```

```python
week = 9
lastweek = week - 1
wk = str(week)
lwk = str(lastweek)

with open('schedule_2018.txt') as f:
    #passer = f.readlines()
    schedule = list(f)

with open('Week' + lwk + '/SRS_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    SRS = list(f)
    
with open('Week' + lwk + '/penalties_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    penalties = list(f)
    
with open('Week' + lwk + '/intercept_pct_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    Intpct = list(f)
    
with open('Week' + lwk + '/TD_pct_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    TDpct = list(f)
    
with open('Week' + lwk + '/sacks_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    sacks = list(f)
    
with open('Week' + lwk + '/rushingDEF_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    rushDEF = list(f)
```

```python
team_list = []
data_dict = {}

for i in range(len(SRS)):
    
    # passing yards per game:
    team = SRS[i][:22].lstrip().rstrip()
    team_list.append(team)
    
    val = float(SRS[i].rstrip()[-7:])
    
    data_dict[team] = {}
    data_dict[team]['SRS'] = val
    
data_dict

    
for i in range(len(SRS)):
    # penalties:
    team = penalties[i][2:25].lstrip()
    
    val = float(penalties[i].rstrip()[-3:])
    
    rnk = int(penalties[i][0:2].rstrip())
    
    data_dict[team]['penalties'] = [val, rnk]
    
    # interception percentage:
    team = Intpct[i][2:25].lstrip()
    
    val = float(Intpct[i].rstrip()[-4:])
    
    rnk = int(Intpct[i][0:2].rstrip())
    
    data_dict[team]['Intpct'] = [val, rnk]
    
    # TD percentage:
    team = TDpct[i][2:25].lstrip()
    
    val = float(TDpct[i].rstrip()[-4:])
    
    rnk = int(TDpct[i][0:2].rstrip())
    
    data_dict[team]['TDpct'] = [val, rnk]
    
    # sack percentage:
    team = sacks[i][2:25].lstrip()
    
    val = float(sacks[i].rstrip()[-4:])
    
    rnk = int(sacks[i][0:2].rstrip())
    
    data_dict[team]['sacks'] = [val, rnk]
    
    # rushing yards allowed per game:
    team = rushDEF[i][2:25].lstrip()
    
    val = float(rushDEF[i].rstrip()[-5:])
    
    rnk = int(rushDEF[i][0:2].rstrip())
    
    data_dict[team]['rushDEF'] = [val, rnk]
    
    
team_list = sorted(team_list)
```

```python
SRS_ar = []

for team in team_list:
    SRS_ar.append([data_dict[team]['SRS'], team])       

power_rankings = sorted(SRS_ar, reverse=True)

print('Power rankings:')
print()
for i in range(32):
    print('{:2d}'.format(i + 1), power_rankings[i][1].rjust(22), '{0:+.1f}'.format(power_rankings[i][0] ))
```

```python
for i in range(len(schedule)):
    if ( schedule[i][0]==wk and schedule[i][1]==' ' ) or schedule[i][0:2]==wk:
        
        team_away = schedule[i][3:29].lstrip().rstrip()
        team_home = schedule[i][28:].lstrip().rstrip()
        
        spread1 = 2.5 + (- data_dict[team_away]['SRS'] + data_dict[team_home]['SRS']) / 2.
        
        rnk_1 = data_dict[team_away]['penalties'][1]
        col_1 = '30'
        if rnk_1 <= 8:
            col_1 = '32'
        elif rnk_1 >= 25:
            col_1 = '31'
        str_1 = str(rnk_1).rjust(3)
            
        rnk_2 = data_dict[team_home]['penalties'][1]
        col_2 = '30'
        if rnk_2 <= 8:
            col_2 = '32'
        elif rnk_2 >= 25:
            col_2 = '31'
        str_2 = str(rnk_2).rjust(3)
        
        rnk_3 = data_dict[team_away]['Intpct'][1]
        col_3 = '30'
        if rnk_3 <= 8:
            col_3 = '32'
        elif rnk_3 >= 25:
            col_3 = '31'
        str_3 = str(rnk_3).rjust(3)
            
        rnk_4 = data_dict[team_home]['Intpct'][1]
        col_4 = '30'
        if rnk_4 <= 8:
            col_4 = '32'
        elif rnk_4 >= 25:
            col_4 = '31'
        str_4 = str(rnk_4).rjust(3)
        
        rnk_5 = data_dict[team_away]['TDpct'][1]
        col_5 = '30'
        if rnk_5 <= 8:
            col_5 = '32'
        elif rnk_5 >= 25:
            col_5 = '31'
        str_5 = str(rnk_5).rjust(3)
            
        rnk_6 = data_dict[team_home]['TDpct'][1]
        col_6 = '30'
        if rnk_6 <= 8:
            col_6 = '32'
        elif rnk_6 >= 25:
            col_6 = '31'
        str_6 = str(rnk_6).rjust(3)
        
        rnk_7 = data_dict[team_away]['sacks'][1]
        col_7 = '30'
        if rnk_7 <= 8:
            col_7 = '32'
        elif rnk_7 >= 25:
            col_7 = '31'
        str_7 = str(rnk_7).rjust(3)
            
        rnk_8 = data_dict[team_home]['sacks'][1]
        col_8 = '30'
        if rnk_8 <= 8:
            col_8 = '32'
        elif rnk_8 >= 25:
            col_8 = '31'
        str_8 = str(rnk_8).rjust(3)
        
        rnk_9 = data_dict[team_away]['rushDEF'][1]
        col_9 = '30'
        if rnk_9 <= 8:
            col_9 = '32'
        elif rnk_9 >= 25:
            col_9 = '31'
        str_9 = str(rnk_9).rjust(3)
            
        rnk_10 = data_dict[team_home]['rushDEF'][1]
        col_10 = '30'
        if rnk_10 <= 8:
            col_10 = '32'
        elif rnk_10 >= 25:
            col_10 = '31'
        str_10 = str(rnk_10).rjust(3)
        
        rnk_away = np.mean([rnk_1, rnk_3, rnk_5, rnk_7, rnk_9])
        rnk_home = np.mean([rnk_2, rnk_4, rnk_6, rnk_8, rnk_10])
        
        spread2 = 2.5 + rnk_away - rnk_home
        
        spread = (spread1 + spread2) / 2.
        
        print(team_away.rjust(22) + ' {0:+5.1f} ({1:+5.1f}, {2:+5.1f})   at   '.format(spread, spread1, spread2) + team_home )
        print()
        #print('penalties:  {0:3d}'.format(data_dict[team_away]['penalties'][1]).rjust(22))
        colorpr2('penalties:    ', str_1, col_1, str_2, col_2)
        colorpr2('Int%:         ', str_3, col_3, str_4, col_4)
        colorpr2('TD%:          ', str_5, col_5, str_6, col_6)
        colorpr2('sacks:        ', str_7, col_7, str_8, col_8)
        colorpr2('rushDEF:      ', str_9, col_9, str_10, col_10)
        print('avg rank:          ', rnk_away, '           ', rnk_home)
        print()
```
